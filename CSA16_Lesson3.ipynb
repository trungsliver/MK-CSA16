# **DATA FRAME: Đọc - ghi file**

1. Đọc ghi file trực tiếp


```python
import pandas as pd
```

```python
# Khởi tạo dữ liệu dạng dictionary
data = {
    'Name': ['Đạt', 'Huy', 'Việt Anh', 'Bình', 'Nghi'],
    'Age': [16, 17, 18, 17, 16],
    'City': ['Hà Nội', 'Hạ Long', 'Hồ Chí Minh', 'Cần Thơ', 'Vũng Tàu']
}
# Khởi tạo DataFrame
df = pd.DataFrame(data)
# Hiển thị dữ liệu từ DataFrame
df
```

Tạo và đọc dữ liệu

```python
# Tạo dữ liệu CSV
df.to_csv('output.csv', index=False)
# Đọc dữ liệu từ file CSV
df_read = pd.read_csv('output.csv')
print(df_read)
```

```python
# Tạo dữ liệu xlsx
df = pd.DataFrame(data)
df.to_excel('output.xlsx', index=False)
# Đọc dữ liệu từ file xlsx
df_read = pd.read_excel('output.xlsx', sheet_name='Sheet1')
print(df_read)
```

```python
# Tạo dữ liệu json
df = pd.DataFrame(data)
df.to_json('output.json', orient='records')
# Đọc dữ liệu từ file json
df_read = pd.read_json('output.json')
print(df_read)
```

# **Truy cập dữ liệu**

Truy cập theo cột

```python
animals = df
# Truy cập theo cột: DataFrame[col]
animals['animal'].head(3)
```

```python
# Truy cập theo cột: DataFrame.col
animals.sound.head(3)
```

```python
# Truy cập nhiều cột: DataFrame[[col1, col2, ...]]
animals[['animal', 'sound']].tail(3)
```

```python
# Biến cột 'animal' thành index để truy cập vào các phần tử
# animals.set_index("animal", inplace=True)

animals.head(3)
```

Truy cập theo hàng

```python
# Lọc các hàng của DataFrame bằng index slicing
    # Hiện 4 hàng đầu tiên
animals[:4]
    # Hiện từ hàng thứ 11 cho đến hết
animals[10:]
    # Hiện từ hàng thứ 5 đến hàng thứ 8
animals[4:8]
```

```python
# Lọc theo cả hàng và cột: DataFrame.iloc[rows, columns]
    # Hiển thị 3 dòng đầu tiên của cột thứ 2
animals.iloc[:3, 1:3]
```

```python
# Lọc dựa trên nhãn
students = pd.DataFrame({
    'name': ['Quang Huy', 'Thành Đạt', 'Việt Anh', 'Ngọc Nghi', 'Gia Bình'],
    'age': [16, 17, 17, 17, 18],
    'gender': ['Male', 'Male', 'Male', 'Female', 'Male']
})

# Chọn name là index
students.set_index('name', inplace=True)
# Hiển thị hàng có index = 'Quang Huy', 'Việt Anh'
students.loc[['Quang Huy', 'Việt Anh'], 'age':'gender']
```

```python
# Lọc kết hợp điều kiện
students.loc[students.age > 17, 'age':'gender']
```
