Xử lý dữ liệu với thư viện pandas

import thư viện

```python
import pandas as pd
```

Series là danh sách một chiều chứa dữ liệu, chỉ số phần tử ở 1 danh sách khác. Chỉ số phần tử có thể là bất kỳ kiểu dữ liệu nào trong numpy.

Ví dụ 1: index mặc định (không khai báo)

```python
names = pd.Series(['Quang Huy', 'Thành Đạt', 'Việt Anh', 'Ngọc Nghi', 'Gia Bình'])
names
```

Series names chứa:
Giá trị (value): Quang Huy, Thành Đạt, ...
Chỉ số phần tử (index): 0,1,2,...
Kiểu dữ liệu của series: object (tương ứng với str)


Ví dụ 2: có khai báo index, có thể không phải kiểu dữ liệu int

```python
names = pd.Series(['Quang Huy', 'Thành Đạt', 'Việt Anh', 'Ngọc Nghi', 'Gia Bình'],
                  index = ['A', 'B', 'C', 'D', 'E'])
names
```

Có thể tạo Series từ dict thay vì list

```python
fruits_dict = {
    'a': 'apple',
    'b': 'banana',
    'c': 'orange',
    'd': 'pear',
    'e': 'strawberry'
}
fruits = pd.Series(fruits_dict)
fruits
```

## Thực hành: Thống kê cân nặng học sinh
Cho các số liệu cân nặng học sinh như sau:
|Tên|Cân nặng|
|---|--------|
|Hoa|37 kg|
|Nam|48 kg|
|Hải|42 kg|
|Linh|39kg|
|Nhân|41kg|
|Phương|38kg|
|Thu|40 kg|
|Vũ|40 kg|
|Bảo|43 kg|

Sử dụng pandas Series, tìm tên các bạn tương ứng với các cân nặng `mean`, `median`, `mode`.

```python
# Khai báo dict
weight_dict = {
    'Hoa': 37,
    'Nam': 48,
    'Hải': 42,
    'Linh': 39,
    'Nhân': 41,
    'Phương': 38,
    'Thu': 40,
    'Vũ': 40,
    'Bảo': 43
}
# Tạo series
weights = pd.Series(weight_dict)
weights
```

```python
# Tìm mean, median, mode
    # Mean (Trung bình): Tổng của tất cả giá trị chia cho số lượng giá trị
mean_weight = weights.mean()
    # Median (Trung vị): Giá trị ở giữa khi series được sắp xếp theo giá trị
median_weight = weights.median()
    # Mode: Giá trị xuất hiện nhiều lần nhất
mode_weight = weights.mode()

# Hiển thị kết quả
print("Cân nặng trung bình:", mean_weight, "kg")
print("Cân nặng trung vị:", median_weight, "kg")
print("Cân nặng phổ biến nhất:", mode_weight, 'kg')
print("Cân nặng phổ biến nhất (mode):")
for m in mode_weight:
    print(m, "kg")
```

Ví dụ: đánh giá chất lượng dịch vụ


```python
ratings = pd.Series([4,3,1,2,1,5,4,1,1])
# Đánh giá
print(f'Chất lượng trung bình: {round(ratings.mean(), 2)}/5')
print('Đánh giá phổ biến nhất: ', end='')
for i in ratings.mode():
    print(i, "sao")
```

DataFrame

DataFrame là cấu trúc dữ liệu dạng 2 chiều (dạng bảng, gồm hàng và cột)

Mỗi cột trong DataFrame là 1 Series


```python
# keys - tương ứng với tên cột
# danh sách values - tương ứng với giá trị của cột đó
students = pd.DataFrame({
    'name': ['Quang Huy', 'Thành Đạt', 'Việt Anh', 'Ngọc Nghi', 'Gia Bình'],
    'age': [16, 17, 17, 17, 18],
    'gender': ['Male', 'Male', 'Male', 'Female', 'Male']
})
students
```

Tóm tắt thông tin DataFrame: DataFrame.info()

```python
students.info()
```

DataFrame.head(n): Hiển thị n giá trị đầu của DataFrame

```python
students.head(3)
```

DataFrame.tail(n): Hiển thị n giá trị cuối của DataFrame

```python
students.tail(3)
```
