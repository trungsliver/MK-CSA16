# **Bin i d liu**

DataFrame.apply():  thay i gi tr ca ct bng 1 hm t nh ngha

```python
df = pd.read_csv("Billionaires.csv")
def clean_networth(networth):
    return networth[1:-1]

# lambda: khai bo hm n danh (khng tn), thng dng  bin i hng lot cc phn t
# x: i din cho mi phn t
# x[1:-1]: ln phn s trong Net Worth (khng ly k t $ v B)
# pd.to_numeric(<ct>): phng thc thay i kiu d liu t string => float

# df["Net Worth"] = df["Net Worth"].apply(lambda x: x[1:-1])
df["Net Worth"] = df["Net Worth"].apply(clean_networth)
df["Net Worth"] = pd.to_numeric(df["Net Worth"])
df.dtypes
```

```python
# Bin i kiu d liu ca age
df["Age"] = pd.to_numeric(df["Age"])
df.dtypes
```

Thao tc nhm d liu

```python
#  bi: "a ra tng ti sn ln nht ca cc t ph nm  lnh vc no"
# => Cn tnh tng lng ti sn theo tng lnh vc

# Tnh tng lng ti sn (Net Worth) theo lnh vc (Industry)
df[['Industry', 'Net Worth']].groupby('Industry').sum()
```

```python
# Lu d liu va tng hp v DataFrame mi
df_industry = df[['Industry', 'Net Worth']].groupby('Industry').sum()

# Sp xp theo gi tr tng dn
df_industry = df_industry.sort_values(by='Net Worth', ascending=False)

# Hin th kt qu
df_industry
```

```python
import pycountry_convert as pc

# Hm tr v tn chu lc t tn quc gia
def get_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name

    except:
        print(country_name)
        return "Unknown"

# To thm ct 'Continent' (chu lc) vo dataFrame
df["Continent"] = df["Country | Territory"].apply(get_continent)
df.head()
```

```python
# Tnh trung bnh ti sn ca t ph theo chu lc
df[['Net Worth', 'Continent']].groupby('Continent').mean()

# Lu d liu va tng hp v DataFrame mi
df_continent = df[['Net Worth', 'Continent']].groupby('Continent').mean()

# Sp xp theo gi tr tng dn
df_continent = df_continent.sort_values(by='Net Worth', ascending=False)

# Hin th kt qu
df_continent
```

```python
# Kim tra quc gia Unknown
df[df["Continent"] == "Unknown"]
```

```python
# Thay th quc gia 'Eswatini (Swaziland)' thnh Swaziland
df["Country | Territory"] = df["Country | Territory"].replace({"Eswatini (Swaziland)":"Swaziland"})
# Chy li hm tm chu lc t tn quc gia
df["Continent"] = df["Country | Territory"].apply(get_continent)
# Tnh li gi tr Net Worth trung bnh ca tng Continent (chu lc)
df[["Net Worth", "Continent"]].groupby("Continent").mean()
```

```python
df = pd.read_csv('hockey_players_data.csv')
df
```

```python
# dropna(): b tt c cc hng c gi tr b khuyt     df.dropna()
# Nu mun b ct, th kt hp vi axis=1               df.dropna(axis=1)
df.dropna()
```

```python
# Loi b 1 s ct khng cn thit: 'birthdate', 'birthplace'
df.drop(['birthdate', 'birthplace'], axis=1, inplace=True)
df
```

```python
# fillna(): thay th cc gi tr NaN thnh gi tr mc nh
    # Thay gi tr NaN thnh 0 trong ct goals
df.goals = df.goals.fillna(0)
df.tail()
```

```python
# X l gi tr sai kiu
df.age = df.age.astype("int")
df.height = pd.to_numeric(df.height)
df.dtypes
```

```python
# Kim tra s lng gi tr b trng
df.duplicated().sum()
# Loi b cc gi tr b trng
df = df.drop_duplicates()
# Kim tra li s lng gi tr b trng
df.duplicated().sum()
```
