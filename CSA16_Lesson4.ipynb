# **Biến đổi dữ liệu**

DataFrame.apply(): để thay đổi giá trị của cột bằng 1 hàm tự định nghĩa

```python
df = pd.read_csv("Billionaires.csv")
def clean_networth(networth):
    return networth[1:-1]

# lambda: khai báo hàm ẩn danh (không tên), thường dùng để biến đổi hàng loạt các phần tử
# x: đại diện cho mỗi phần tử
# x[1:-1]: lần phần số trong Net Worth (không lấy ký tự $ và B)
# pd.to_numeric(<cột>): phương thức thay đổi kiểu dữ liệu từ string => float

# df["Net Worth"] = df["Net Worth"].apply(lambda x: x[1:-1])
df["Net Worth"] = df["Net Worth"].apply(clean_networth)
df["Net Worth"] = pd.to_numeric(df["Net Worth"])
df.dtypes
```

```python
# Biến đổi kiểu dữ liệu của age
df["Age"] = pd.to_numeric(df["Age"])
df.dtypes
```

# **Nhóm và sắp xếp**

Thao tác nhóm dữ liệu

```python
# Đề bài: "Đưa ra tổng tài sản lớn nhất của các tỉ phú nằm ở lĩnh vực nào"
# => Cần tính tổng lượng tài sản theo từng lĩnh vực

# Tính tổng lượng tài sản (Net Worth) theo lĩnh vực (Industry)
df[['Industry', 'Net Worth']].groupby('Industry').sum()
```

Thao tác sắp xếp

```python
# Lưu dữ liệu vừa tổng hợp và DataFrame mới
df_industry = df[['Industry', 'Net Worth']].groupby('Industry').sum()

# Sắp xếp theo giá trị tăng dần
df_industry = df_industry.sort_values(by='Net Worth', ascending=False)

# Hiển thị kết quả
df_industry
```

Bài tập thực hành: Tính trung bình tài sản của tỉ phú theo châu lục

```python
import pycountry_convert as pc

# Hàm trả về tên châu lục từ tên quốc gia
def get_continent(country_name):
    try:
        country_code = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_code)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name

    except:
        print(country_name)
        return "Unknown"

# Tạo thêm cột 'Continent' (châu lục) vào dataFrame
df["Continent"] = df["Country | Territory"].apply(get_continent)
df.head()
```

```python
# Tính trung bình tài sản của tỉ phú theo châu lục
df[['Net Worth', 'Continent']].groupby('Continent').mean()

# Lưu dữ liệu vừa tổng hợp và DataFrame mới
df_continent = df[['Net Worth', 'Continent']].groupby('Continent').mean()

# Sắp xếp theo giá trị tăng dần
df_continent = df_continent.sort_values(by='Net Worth', ascending=False)

# Hiển thị kết quả
df_continent
```

```python
# Kiểm tra quốc gia Unknown
df[df["Continent"] == "Unknown"]
```

```python
# Thay thế quốc gia 'Eswatini (Swaziland)' thành Swaziland
df["Country | Territory"] = df["Country | Territory"].replace({"Eswatini (Swaziland)":"Swaziland"})
# Chạy lại hàm tìm châu lục từ tên quốc gia
df["Continent"] = df["Country | Territory"].apply(get_continent)
# Tính lại giá trị Net Worth trung bình của từng Continent (châu lục)
df[["Net Worth", "Continent"]].groupby("Continent").mean()
```

# **Data Cleaning - Làm sạch dữ liệu**

```python
df = pd.read_csv('hockey_players_data.csv')
df
```

```python
# Đếm số giá trị bị khuyết
df.isnull().sum()
```

```python
# dropna(): bỏ tất cả các hàng có giá trị bị khuyết     df.dropna()
# Nếu muốn bỏ cột, thì kết hợp với axis=1               df.dropna(axis=1)
df.dropna()
```

```python
# Loại bỏ 1 số cột không cần thiết: 'birthdate', 'birthplace'
df.drop(['birthdate', 'birthplace'], axis=1, inplace=True)
df
```

```python
# fillna(): thay thế các giá trị NaN thành giá trị mặc định
    # Thay giá trị NaN thành 0 trong cột goals
df.goals = df.goals.fillna(0)
df.tail()
```

```python
# Xử lý giá trị sai kiểu
df.age = df.age.astype("int")
df.height = pd.to_numeric(df.height)
df.dtypes
```

```python
# Kiểm tra số lượng giá trị bị trùng
df.duplicated().sum()
# Loại bỏ các giá trị bị trùng
df = df.drop_duplicates()
# Kiểm tra lại số lượng giá trị bị trùng
df.duplicated().sum()
```
